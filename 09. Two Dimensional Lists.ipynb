```python
from sys import stdin

def rowWiseSum(mat, nRows, mCols):
    for i in range(nRows):
        sum=0
        for j in range(mCols):
            sum+=mat[i][j]
        print(sum,end=' ')    

#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    rowWiseSum(mat, nRows, mCols)
    print()

    t -= 1
```

```python
from sys import stdin

def spiralPrint(mat,m,n):
    
    rs, re = 0, n # rs-> row start, re-> row end
    cs, ce = 0, m #cs-> column start, ce-> column end
    count = 0 
    while (count < (n*m)): # total n*m elements need to be printed
        # left to right top
        i = rs 
        for j in range(cs,ce):
            print(mat[i][j], end=' ')
            count += 1
        
        rs += 1 # updating 
        j = ce-1
        for i in range(rs,re):
            print(mat[i][j], end=' ')
            count += 1
        ce -= 1
        i = re-1
        for j in range(ce-1,cs-1,-1):
            print(mat[i][j], end=' ')
            count += 1
        re -= 1
        j = cs
        for i in range(re-1,rs-1,-1):
            print(mat[i][j], end=' ')
            count += 1
        cs += 1
            
#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    n = int(li[0])
    m = int(li[1])
    if n == 0 :
        return list(), 0, 0
    mat = [list(map(int, input().strip().split(" "))) for row in range(n)]
    return mat, n, m
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, n, m = take2DInput()
    spiralPrint(mat,m,n)
    print()

    t -= 1
```
